```python
import os

def disable_ssl():
    # Fix SSL certificate verification issues on Windows
    import ssl
    import warnings
    warnings.filterwarnings('ignore', message='Unverified HTTPS request')

    # Disable SSL verification globally
    ssl._create_default_https_context = ssl._create_unverified_context

    # Patch requests to disable SSL verification
    import requests
    from functools import partialmethod
    requests.Session.request = partialmethod(requests.Session.request, verify=False)
    requests.request = partialmethod(requests.request, verify=False)
```

```python
from PIL import Image
from surya.foundation import FoundationPredictor
from surya.recognition import RecognitionPredictor
from surya.detection import DetectionPredictor

disable_ssl()
image = Image.open(r"C:\Users\I1000929\Downloads\Smart-Licence-Card-1.jpg")

foundation_predictor = FoundationPredictor()
recognition_predictor = RecognitionPredictor(foundation_predictor)
detection_predictor = DetectionPredictor()

predictions = recognition_predictor([image], det_predictor=detection_predictor)
```

```python
output=""
for result in predictions:   # ocr_results = list[OCRResult]
    for idx, line in enumerate(result.text_lines, start=1):
        output += f"{line.text} | (confidence={line.confidence:.2f})\n"

print(output)
```

```python
import cv2
import numpy as np
from PIL import Image


def preprocess_image_for_ocr(
    pil_img: Image.Image,
    max_width: int = 1200
) -> Image.Image:
    """
    Reduce image size and preprocess to keep only readable text for OCR.
    """

    # ---------------------------
    # PIL → OpenCV
    # ---------------------------
    img = np.array(pil_img)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    # ---------------------------
    # Resize (maintain aspect ratio)
    # ---------------------------
    h, w = img.shape[:2]
    if w > max_width:
        scale = max_width / w
        img = cv2.resize(img, (int(w * scale), int(h * scale)),
                         interpolation=cv2.INTER_AREA)

    # ---------------------------
    # Convert to grayscale
    # ---------------------------
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # ---------------------------
    # Contrast enhancement (CLAHE)
    # ---------------------------
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    gray = clahe.apply(gray)

    # ---------------------------
    # Noise removal (keeps text thin)
    # ---------------------------
    gray = cv2.bilateralFilter(gray, d=9, sigmaColor=75, sigmaSpace=75)

    # ---------------------------
    # Adaptive threshold → text only
    # ---------------------------
    thresh = cv2.adaptiveThreshold(
        gray,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        31,
        10
    )

    # ---------------------------
    # Remove tiny noise (optional)
    # ---------------------------
    kernel = np.ones((2, 2), np.uint8)
    cleaned = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

    # ---------------------------
    # OpenCV → PIL
    # ---------------------------
    return Image.fromarray(cleaned)
```

```python
from PIL import Image

img = Image.open("C:\\yy\\JJ.jpg")
processed_img = preprocess_image_for_ocr(img)

processed_img.save("processed_for_ocr.png")
```
